```python
from delta.tables import DeltaTable, IdentityGenerator
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

# **Create Delta Table**

## SQL API


```python
%sql
CREATE TABLE deltalakecesar.default.first_table
(
  id INT,
  salary INT
)
USING DELTA 
```

#### Using Delta API

```python
# Create table in the metastore
DeltaTable.createIfNotExists(spark) \
  .tableName("deltalakecesar.default.firstdeltaapi") \
  .addColumn("id", "INT") \
  .addColumn("salary", "INT") \
  .execute()
```

```python
%sql 
DROP TABLE deltalakecesar.default.firstdeltaapi
```

# **GENERATED COLUMNS**

#### **Identity Columns**

```python
DeltaTable.create(spark)\
  .tableName("deltalakecesar.default.firstdeltaapi")\
  .addColumn("id_col", dataType=LongType(), generatedAlwaysAs=IdentityGenerator()) \
  .addColumn("salary", dataType=IntegerType()) \
  .addColumn("name", dataType=StringType())\
  .execute()
```

#### **Computed Columns**

```python
DeltaTable.create(spark)\
  .tableName("deltalakecesar.default.firstdeltaapi2")\
  .addColumn("salaryAfterTax", dataType=LongType(), generatedAlwaysAs="CAST((salary * 0.8) AS BIGINT)") \
  .addColumn("salary", dataType=IntegerType()) \
  .addColumn("name", dataType=StringType())\
  .execute()
```

```python
%sql
INSERT INTO deltalakecesar.default.firstdeltaapi2(salary,name)
VALUES (100, 'alice'),
       (200, 'bob'),
       (300, 'charlie')
```
