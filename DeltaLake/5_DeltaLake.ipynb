# DELTA LOG LEVEL SCHEMA CHANGES

```python
data = [(1,100,'xyz',1),(9,200,'bb',1),(10,300,'cc',1)] 
df = spark.createDataFrame(data, ['id','income','name','tip'])
display(df)
```

```python
df.write.format("delta").mode("append").save("/Volumes/deltalakecesar/default/deltavolume/schemalevel/")
```

 if we want to change column will change but not the schema but will change in the delta log


```python

%sql

ALTER TABLE delta.`/Volumes/deltalakecesar/default/deltavolume/schemalevel/`
SET TBLPROPERTIES (
  delta.minReadVersion = '2',
  delta.minWriterVersion = '5',
  delta.columnMapping.mode = 'name'
);
```
