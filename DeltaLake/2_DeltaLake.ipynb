```python
data = [(1,100,'aa'),(2,200,'bb'),(3,300,'cc'),(4,400,'dd')]
df = spark.createDataFrame(data, ['id','salary','name'])

display(df)
```

```python
df.write.format("delta")\
.mode("append")\
.save("/Volumes/deltalakecesar/default/deltavolume/demosink/")
```
