# **SCHEMA ENFORCEMENT**

make sure your data is consistent


```python
data = [(1,100,'aa',1),(2,200,'bb',1),(3,300,'cc',1),(4,400,'dd',1)]
df = spark.createDataFrame(data, ['id','salary','name'])

display(df)
```

```python
df.write.format("delta")\
.mode("append")\
.save("/Volumes/deltalakecesar/default/deltavolume/demosink/")
```

know imagine you waant to evolve the schema

# **SCHEMA EVOLUTION**

```python
df.write.format("delta").option("mergeSchema", "true")\
.mode("append")\
.option("mergeSchema", True)\
.save("/Volumes/deltalakecesar/default/deltavolume/demosink/")
```

# SCHEMA OVERWRITE

```python
data = [(1,100,'aa',1),(2,200,'bb',1),(3,300,'cc',1),(4,400,'dd',1)]
df = spark.createDataFrame(data, ['id','income','name','tip'])

display(df)
```

```python
df.write.format("delta")\
.mode("overwrite")\
.option("overwriteSchema", True)\
.save("/Volumes/deltalakecesar/default/deltavolume/demosink/")
```
