# Optimization

too many files, small file probs



small file problem occurs when a table is composed of thousands or millions of tiny files (often just a few KBs each) rather than a smaller number of large, healthy files (around 128MB+). This creates a massive performance bottleneck because the Spark driver becomes overwhelmed keeping track of all the file locations, and the system spends more time opening and closing files

to solve optinzeit will become lesser and bigger files using coalesce


```python
df = spark.read.table("deltalakecesar.default.clonetbl")
```

```python
display(df)
```

```python
df.write.format("delta")\
    .mode("overwrite")\
    .save("/Volumes/deltalakecesar/default/deltavolume/Optimization/")
```

```python
%sql
OPTIMIZE delta.`/Volumes/deltalakecesar/default/deltavolume/Optimization/`
```

```python
df.write.format("delta")\
    .mode("append")\
    .save("/Volumes/deltalakecesar/default/deltavolume/Optimization/")
```

```python
%sql
OPTIMIZE delta.`/Volumes/deltalakecesar/default/deltavolume/Optimization/`
```

```python
%sql
DESCRIBE HISTORY delta.`/Volumes/deltalakecesar/default/deltavolume/Optimization/`
```

### Zordering
 create big partition where sort similar data in the partitions and calculate cols stats. this is the concept of data skiopping

```python
%sql
SELECT * FROM delta.`/Volumes/deltalakecesar/default/deltavolume/Optimization/`
WHERE id = 1
```

```python
%sql
OPTIMIZE delta.`/Volumes/deltalakecesar/default/deltavolume/Optimization/` ZORDER BY (id)
```

```python
%sql
SELECT * FROM delta.`/Volumes/deltalakecesar/default/deltavolume/Optimization/`
WHERE id = 1
```
